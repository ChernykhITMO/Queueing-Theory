# **Решение**

## 1. Описание модели

Рассматривается многоканальная система массового обслуживания типа $M/M/n/\infty$

Характеристики системы:
- входной поток заявок является пуассоновским с интенсивностью $\lambda$
- время обслуживания имеет экспоненциальное распределение с параметром $\mu$
- система содержит $n$ параллельных каналов обслуживания
- длина очереди не ограничена
- заявки обслуживаются в порядке поступления (FIFO)

Если все каналы заняты, заявка становится в очередь и ожидает освобождения канала

## 2. Параметры системы

В качестве единицы времени выбран **час**.

Заданы следующие параметры системы:

- интенсивность входящего потока:
  $\lambda = 10$ заявок/час;

- интенсивность обслуживания одного канала:
  $\mu = 3$ заявки/час;

- количество каналов обслуживания:
  $n = 4$;

Таким образом, каждый канал в среднем обслуживает одну заявку за
$\frac{1}{\mu} = \frac{1}{3} \approx 0.333$ часа.

## 3. Алгоритм расчета характеристик системы

Расчет характеристик системы выполняется в следующей последовательности:

1. Вычисляется коэффициент загрузки системы:
   $\rho = \frac{\lambda}{n \mu}$

2. Проверяется условие устойчивости:
   если $\rho < 1$, система устойчива и расчеты возможны

3. Вычисляется вероятность простоя системы $P_0$

4. Определяется вероятность ожидания заявки (формула Эрланга C)

5. Рассчитываются основные характеристики:
   - среднее число заявок в очереди $L_q$
   - среднее время ожидания $W_q$
   - среднее время пребывания в системе $W$
   - среднее число заявок в системе $L$

## 4. Формулы, используемые для расчёта
$$
\lambda \text{ — интенсивность входящего потока (заявок/час)}
$$

$$
\mu \text{ — интенсивность обслуживания (заявок/час)}
$$

$$
n \text{ — число каналов обслуживания}
$$

$$
a = \frac{\lambda}{\mu} \text{ — приведённая нагрузка}
$$

$$
\rho = \frac{\lambda}{n \mu} \text{ — коэффициент загрузки системы}
$$

---

## Условие устойчивости системы

$$
\rho < 1
$$

---

## Вероятность пустой системы

$$
P_0 = \left( \sum_{k=0}^{n-1} \frac{a^k}{k!} + \frac{a^n}{n!} \cdot \frac{1}{1 - \rho} \right)^{-1}
$$

---

## Средняя длина очереди

$$
L_q = \frac{a^n \rho}{n! (1 - \rho)^2} \cdot P_0
$$

---

## Среднее число заявок в системе

$$
L = L_q + \frac{\lambda}{\mu}
$$

---

## Среднее время ожидания в очереди

$$
W_q = \frac{L_q}{\lambda}
$$

---

## Среднее время пребывания заявки в системе

$$
W = W_q + \frac{1}{\mu}
$$

---

## Вероятность ожидания (формула Эрланга C)

$$
P_{wait} = \frac{a^n}{n! (1 - \rho)} \cdot P_0
$$

---

## Закон Литтла

$$
L = \lambda W
$$

$$
L_q = \lambda W_q
$$

```python
import math

def factorial(n):
    return math.factorial(n)


def calc_metrics(lmbda, mu, n):
    rho = lmbda / (n * mu)
    if rho >= 1:
        return None

    a = lmbda / mu

    sum_part = 0
    for k in range(n):
        sum_part += (a ** k) / factorial(k)

    last_part = (a ** n) / (factorial(n) * (1 - rho))
    p0 = 1 / (sum_part + last_part)

    p_wait = (a ** n) / (factorial(n) * (1 - rho)) * p0

    Lq = (p_wait * rho) / (1 - rho)
    Wq = Lq / lmbda
    W = Wq + 1 / mu
    L = lmbda * W

    return rho, p0, p_wait, Lq, Wq, W, L


lmbda = 10
mu = 3

print(f"{'n':>2} | {'rho':>6} | {'P0':>6} | {'P_wait':>6} | {'Lq':>6} | {'Wq':>6} | {'W':>6} | {'L':>6}")
print("-" * 70)

for n in range(1, 11):
    res = calc_metrics(lmbda, mu, n)

    if res is None:
        print(f"{n:2} | {'-':>6} | {'-':>6} | {'-':>6} | {'-':>6} | {'-':>6} | {'-':>6} | {'-':>6}")
        continue

    rho, p0, p_wait, Lq, Wq, W, L = res

    print(f"{n:2} | {rho:6.3f} | {p0:6.3f} | {p_wait:6.3f} | {Lq:6.3f} | {Wq:6.3f} | {W:6.3f} | {L:6.3f}")
```


Система массового обслуживания является устойчивой, если коэффициент загрузки ρ < 1

Если ρ ≥ 1, интенсивность поступления заявок превышает суммарную интенсивность обслуживания, что приводит к неограниченному росту очереди

```python
import random
import math
import matplotlib.pyplot as plt
from collections import deque


def simulate(lmbda=10, mu=3, n=4, T=10000):
    current_time = 0.0
    next_arrival = random.expovariate(lmbda)

    service_end_times = []
    queue = deque()

    total_waiting_time = 0.0
    area_under_queue = 0.0
    last_event_time = 0.0

    serviced = 0
    arrived = 0
    waited = 0

    while current_time < T:
        next_service = min(service_end_times) if service_end_times else float('inf')

        if next_arrival < next_service:
            current_time = next_arrival

            area_under_queue += len(queue) * (current_time - last_event_time)
            last_event_time = current_time

            arrived += 1

            if len(service_end_times) < n:
                service_time = random.expovariate(mu)
                service_end_times.append(current_time + service_time)
            else:
                queue.append(current_time)
                waited += 1

            next_arrival = current_time + random.expovariate(lmbda)

        else:
            current_time = next_service

            area_under_queue += len(queue) * (current_time - last_event_time)
            last_event_time = current_time

            service_end_times.remove(next_service)
            serviced += 1

            if queue:
                arrival_time = queue.popleft()
                total_waiting_time += current_time - arrival_time

                service_time = random.expovariate(mu)
                service_end_times.append(current_time + service_time)

    Lq = area_under_queue / T
    Wq = total_waiting_time / serviced if serviced else 0 
    W = Wq + 1 / mu
    L = lmbda * W
    p_wait = waited / arrived if arrived else 0

    return {
        "Lq": Lq,
        "Wq": Wq,
        "W": W,
        "L": L,
        "p_wait": p_wait,
        "arrived": arrived,
        "serviced": serviced
    }

res = simulate(lmbda=10, mu=3, n=4)

print(f"Пришло заявок: {res['arrived']}")
print(f"Обслужено: {res['serviced']}")

print(f"Lq (длина очереди): {res['Lq']:.4f}")
print(f"Wq (время ожидания): {res['Wq']:.4f}")
print(f"W  (время в системе): {res['W']:.4f}")
print(f"L  (заявок в системе): {res['L']:.4f}")
print(f"P_wait (вероятность ожидания): {res['p_wait']:.4f}")
```

```python
import matplotlib.pyplot as plt

n_values = range(1, 15)

Lq_theory = []
Wq_theory = []

Lq_sim = []
Wq_sim = []

valid_n = []

runs = 5

for n in n_values:
    m = calc_metrics(lmbda, mu, n)

    if m:
        rho, p0, p_wait, Lq, Wq, W, L = m

        Lq_theory.append(Lq)
        Wq_theory.append(Wq)
        sim_Lq = 0
        sim_Wq = 0

        for _ in range(runs):
            sim = simulate(lmbda, mu, n)
            sim_Lq += sim["Lq"]
            sim_Wq += sim["Wq"]

        sim_Lq /= runs
        sim_Wq /= runs

        Lq_sim.append(sim_Lq)
        Wq_sim.append(sim_Wq)

        valid_n.append(n)

plt.figure()
plt.plot(valid_n, Wq_theory, marker='o', label='Теория')
plt.plot(valid_n, Wq_sim, marker='x', linestyle='--', label='Моделирование')
plt.title("Wq: теория vs моделирование")
plt.xlabel("Количество каналов")
plt.ylabel("Время ожидания")
plt.legend()
plt.grid()
plt.show()
plt.figure()
plt.plot(valid_n, Lq_theory, marker='o', label='Теория')
plt.plot(valid_n, Lq_sim, marker='x', linestyle='--', label='Моделирование')
plt.title("Lq: теория vs моделирование")
plt.xlabel("Количество каналов")
plt.ylabel("Длина очереди")
plt.legend()
plt.grid()
plt.show()
```

## Анализ результатов

При увеличении числа каналов обслуживания система переходит из неустойчивого состояния (n = 1–3) в устойчивое при n ≥ 4, так как выполняется условие ρ < 1

При n = 4 система уже устойчива, однако наблюдается значительное время ожидания и длина очереди, что свидетельствует о высокой загрузке каналов

При увеличении числа каналов до n = 5–6 происходит резкое снижение времени ожидания Wq и длины очереди Lq, что говорит о существенном улучшении работы системы.

Дальнейшее увеличение числа каналов (n ≥ 7) приводит к незначительным изменениям характеристик. Это указывает на достижение насыщения системы

Таким образом, оптимальное число каналов обслуживания находится в диапазоне 5–6, при котором достигается баланс между эффективностью обслуживания и затратами ресурсов
